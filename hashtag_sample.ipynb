```python
import pandas as pd
df = pd.read_csv("hashtag_201817.csv")
df.head()
t = []
with open("hashtag_201817.csv",encoding='utf8', errors='ignore') as f:
    mainlist=[]
    for line in f:
        temp=line.split(",")
        sublist=[]
        for word in temp:
            word=word.replace("'","")
            word=word.replace('"',"")
            word=word.replace("\n","")
            if word and word!='\n' and word.startswith("#"):
                sublist.append(word)
            if word and word.startswith("#"):
                t.append(word)
        mainlist.append(sublist)
```

```python
hashtags=list(set(t))
```

```python
#print(hashtags)
```

```python
dictionary={}
list1=[]
for i,tag in enumerate(hashtags):
    c=0
    for w in t:
        if tag.lower()==w.lower():
            c=c+1
    if tag not in list1:
        list1.append(tag)
    dictionary[tag]=c
```
